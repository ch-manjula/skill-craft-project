```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(r"C:\Users\Administrator\Downloads\world_population.csv")

# Extract the population data for visualization
population_data = df["2022 Population"]

# Create histogram
plt.figure(figsize=(10,6))
plt.hist(population_data, bins=20, edgecolor='black')  # Divide into 20 bins
plt.xlabel("Population Size")
plt.ylabel("Number of Countries")
plt.title("World Population Distribution (2022)")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv(r"C:\Users\Administrator\Downloads\world_population.csv")

# Select the top 10 most populated countries
top_countries = df.nlargest(10, "2022 Population")

# Create bar chart
plt.figure(figsize=(12,6))
plt.bar(top_countries["Country/Territory"], top_countries["2022 Population"], color='skyblue')
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2022)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
